```python
def intcode(program, inlist=[], i=0):
    outbuf = []
    while True:
        if i >= len(program):
            print('end of program')
            return program
        opcode = program[i]
        
#         print(opcode, program)
        
        op = opcode % 100
        argmodes = []
        argcode = opcode // 100
        while argcode > 0:
            argmodes.append(argcode % 10)
            argcode = argcode // 10
            
#         print(argmodes)
#         print('---')
            
        # ADD A B -> C
        if op == 1:
            if len(argmodes) < 1 or argmodes[0] == 0:
                arg1 = program[program[i+1]]
            else:
                arg1 = program[i+1]
            
            if len(argmodes) < 2 or argmodes[1] == 0:
                arg2 = program[program[i+2]]
            else:
                arg2 = program[i+2]
                
            program[program[i+3]] = arg1 + arg2
            i += 4
        # MULTIPLY A B ->
        elif op == 2:
            if len(argmodes) < 1 or argmodes[0] == 0:
                arg1 = program[program[i+1]]
            else:
                arg1 = program[i+1]
            
            if len(argmodes) < 2 or argmodes[1] == 0:
                arg2 = program[program[i+2]]
            else:
                arg2 = program[i+2]
                
            program[program[i+3]] = arg1 * arg2
            i += 4
        # INPUT -> A
        elif op == 3:
            program[program[i+1]] = inlist[0]
            inlist = inlist[1:]
            i += 2
        # OUTPUT A
        elif op == 4:
#             print('out at', i, program[i+1], program[program[i+1]])
            outbuf.append(program[program[i+1]])
            return program[program[i+1]], i+2, program
            i += 2
        # JUMP TRUE A != 0 -> B
        elif op == 5:
            if len(argmodes) < 1 or argmodes[0] == 0:
                arg1 = program[program[i+1]]
            else:
                arg1 = program[i+1]
                
            if len(argmodes) < 2 or argmodes[1] == 0:
                arg2 = program[program[i+2]]
            else:
                arg2 = program[i+2]
                
            if arg1 != 0:
                i = arg2
            else:
                i += 3
        # JUMP FALSE A == 0 -> B
        elif op == 6:
            if len(argmodes) < 1 or argmodes[0] == 0:
                arg1 = program[program[i+1]]
            else:
                arg1 = program[i+1]
                
            if len(argmodes) < 2 or argmodes[1] == 0:
                arg2 = program[program[i+2]]
            else:
                arg2 = program[i+2]
                
            if arg1 == 0:
                i = arg2
            else:
                i += 3
        # LESS THAN A < B 1 -> C
        elif op == 7:
            if len(argmodes) < 1 or argmodes[0] == 0:
                arg1 = program[program[i+1]]
            else:
                arg1 = program[i+1]
                
            if len(argmodes) < 2 or argmodes[1] == 0:
                arg2 = program[program[i+2]]
            else:
                arg2 = program[i+2]
                
            if arg1 < arg2:
                program[program[i+3]] = 1
            else:
                program[program[i+3]] = 0
            i += 4
        # EQUALS A == B 1 -> C
        elif op == 8:
            if len(argmodes) < 1 or argmodes[0] == 0:
                arg1 = program[program[i+1]]
            else:
                arg1 = program[i+1]
                
            if len(argmodes) < 2 or argmodes[1] == 0:
                arg2 = program[program[i+2]]
            else:
                arg2 = program[i+2]
                
            if arg1 == arg2:
                program[program[i+3]] = 1
            else:
                program[program[i+3]] = 0
            i += 4
        # HALT
        elif op == 99:
            return -1, i, program
        # INVALID
        else:
            print(program[i], 'is not an instruction')
            return -1, i, program
```

```python
def counto(x):
    count = 0
    while count < x:
        yield x, count
        count += 1
```

```python
counter = counto(10)
for c in counter:
    print(c)
```
