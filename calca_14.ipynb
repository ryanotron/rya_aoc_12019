```python
def parse_reactstr(reactstr):
    reactions = {}
    inventory = {}
    for row in reactstr.split('\n'):
        inputstr, outputstr = row.split("=>")
        
        inputs = []
        for elem in inputstr.split(','):
            quant, molec = elem.split()
            inputs.append((int(quant.strip()), molec.strip()))
            
        outquant, outmolec = outputstr.split()
        reactions[outmolec.strip()] = {'get': int(outquant.strip()), 'need': inputs}
        inventory[outmolec.strip()] = 0
    return reactions, inventory

def printstock(inventory):
    for inv in inventory.keys():
        if inventory[inv] > 0:
            print('-', inv, inventory[inv])

def get_reqs(reactions, inventory, target='FUEL', quant=1, prefix=''):
    if target == 'ORE':
#         printstock(inventory)
        return quant, inventory
    
#     print(prefix, 'synthesizing', quant, target)
#     print(prefix, 'reaction: ', reactions[target]['get'], target, '<=', reactions[target]['need'])
    
    newquant = max(0, quant - inventory[target])
    inventory[target] = max(0, inventory[target]-quant)
    quant = newquant

    if quant == 0:
#         print('we have enough {} ({}) in inventory'.format(target, inventory[target]))
#         printstock(inventory)
        return 0, inventory
    
    needs = reactions[target]['need']
    get = reactions[target]['get']
    N = math.ceil(quant/get)
        
#     for need in needs:
#         print(prefix, 'need {} x {} = {} of {}'.format(N, need[0], N*need[0], need[1]))
#     print(prefix, '---')
    
    orecount = 0
    for need in needs:
        extraore, inventory = get_reqs(reactions, inventory, need[1], N*need[0], prefix=prefix+'--')
        orecount += extraore
    inventory[target] += N*get - quant
    
#     if inventory[target] > 0:
#         print(target, inventory[target])
#     else:
#         print('no leftover', target)
#         print('---')
#     printstock(inventory)
#     print(prefix, '--- end synth {}'.format(target))
    return orecount, inventory
    
```

```python
from copy import deepcopy
```

```python
with open('calca_14_input.txt', 'r') as reactfile:
    reactstr = ''
    for line in reactfile:
        reactstr += line
        
reactions, inventory = parse_reactstr(reactstr)   
oc, invx = get_reqs(reactions, inventory, 'FUEL', 1)
print(oc)

# maxore = 1e12
maxore = 201324
fuelcount = 0
step = 1000000
reactions, inventory = parse_reactstr(reactstr)
while maxore > 0:
    oc, newinventory = get_reqs(reactions, deepcopy(inventory), 'FUEL', step)
    print(step, oc)
    if maxore - oc >= 0:
        maxore -= oc
        fuelcount += step
        inventory = deepcopy(newinventory)
        print('intermediate', fuelcount, maxore)
    else:
        step = step // 10
        print('step', step)
        if step == 0:
            break
print(fuelcount)
```
