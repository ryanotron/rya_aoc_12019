```python
def deal_new(deck):
    newdeck = deck.copy()
    newdeck.reverse()
    return newdeck

def cut_N(deck, N):
    newdeck = deck[N:] + deck[:N]
    return newdeck

def deal_inc(deck, inc):
    newdeck = [0]*len(deck)
    j = 0
    for i in range(len(deck)):
        newdeck[j] = deck[i]
        j += inc
        j = (j%len(deck))
    return newdeck
    
```

```python
def intercmd(cmdlist, decksize=10):
    deck = list(range(decksize))
    for cmd in cmdlist:
        if 'deal into new stack' in cmd:
#             print('deal into new stack')
            deck = deal_new(deck)
        elif 'cut' in cmd:
            N = int(cmd.strip().split()[-1])
#             print('cut', N)
            deck = cut_N(deck, N)
        elif 'deal with increment' in cmd:
            inc = int(cmd.strip().split()[-1])
#             print('deal with increment', inc)
            deck = deal_inc(deck, inc)
    return deck
```

```python
with open('calca_22_input.txt', 'r') as infile:
    cmdlist = [row.strip() for row in infile if len(row) > 1]
    
newdeck = intercmd(cmdlist, decksize=10007)
for i, card in enumerate(newdeck):
    if card == 2019:
        print(i, card)
        break
```

```python
def pidx_deal_new(idx, dsize):
    return dsize - 1 - idx

def pidx_cut_N(idx, N, dsize):
    pidx = (idx + N + dsize) % dsize
    return pidx

def pidx_deal_inc(idx, inc, dsize):
    c = 0
    rem_i = (idx + c*dsize)%inc
    while not (rem_i == 0):
        c += 1
        rem_i = (idx + c*dsize)%inc
    return (idx + c*dsize)//inc
```

```python
def rev_inter(revcmd, idx, dsize=10):
    for cmd in revcmd:
        if 'deal into new stack' in cmd:
#             print('deal into new stack')
            idx = pidx_deal_new(idx, dsize)
        elif 'cut' in cmd:
            N = int(cmd.strip().split()[-1])
#             print('cut', N)
            idx = pidx_cut_N(idx, N, dsize)
        elif 'deal with increment' in cmd:
            inc = int(cmd.strip().split()[-1])
#             print('deal with increment', inc)
            idx = pidx_deal_inc(idx, inc, dsize)
    return idx
```
