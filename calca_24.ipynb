```python
from copy import deepcopy
```

```python
def parsegrid(rawgrid):
    return [[1 if col=='#' else 0 for col in row] for row in rawgrid]

def drawgrid(grid):
    strgrid = '\n'.join([' '.join(['.' if col==0 else '#' for col in row]) for row in grid])
    print(strgrid)
    
def zerogrid(row=5, col=5):
    grid = []
    for i in range(row):
        rowlist = []
        for j in range(col):
            rowlist.append(0)
        grid.append(rowlist)
    return grid

def step(grid):
    adjacents = [[0, 1], [0, -1], [1, 0], [-1, 0]]
    nextgrid = deepcopy(grid)
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            nsum = 0
            for adj in adjacents:
                ni = i + adj[0]
                nj = j + adj[1]
                
                if ni >= 0 and nj >= 0 and ni < len(grid) and nj < len(grid[0]):
                    nsum += grid[ni][nj]
                    
            if grid[i][j] == 1 and (not nsum == 1):
                nextgrid[i][j] = 0
                
            elif grid[i][j] == 0 and (nsum == 1 or nsum == 2):
                nextgrid[i][j] = 1
                
    return nextgrid

def step_recur(grids):
    adjacents = [[0, 1], [0, -1], [1, 0], [-1, 0]]
    nextgrids = deepcopy(grids)
    
    levels = list(grids.keys())
    newlevels = []
    for glevel in levels:
        for i in range(len(grids[glevel])):
            for j in range(len(grids[glevel][0])):
                nsum = 0
                
                if i == 2 and j == 2:
                    continue
                
                for adj in adjacents:
                    ni = i + adj[0]
                    nj = j + adj[1]
#                     if glevel == 0 and i == 0 and j == 1:
#                         print(ni, nj)
                
                    if ni == 2 and nj == 2:
                        lower = glevel + 1
                        if lower in grids:
                            if i == 1:
                                for lj in range(len(grids[lower][0])):
                                    nsum += grids[lower][0][lj]
                            elif i == 3:
                                for lj in range(len(grids[lower][0])):
                                    nsum += grids[lower][-1][lj]
                            elif j == 1:
                                for li in range(len(grids[lower])):
                                    nsum += grids[lower][li][0]
                            elif j == 3:
                                for li in range(len(grids[lower])):
                                    nsum += grids[lower][li][-1]
                        else:
                            grids[lower] = zerogrid()
                            nextgrids[lower] = zerogrid()
                            newlevels.append(lower)

                    elif ni >= 0 and nj >= 0 and ni < len(grids[glevel]) and nj < len(grids[glevel][0]):
                        nsum += grids[glevel][ni][nj]
#                         if glevel == 0 and i == 0 and j == 1:
#                             print('level adjacent', ni, nj, nsum)

                    elif ni < 0:
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][1][2]
#                             if glevel == 0 and i == 0 and j == 1:
#                                 print('higher adjacent', ni, nj, nsum)
                        else:
                            grids[higher] = zerogrid()
                            nextgrids[higher] = zerogrid()
                            newlevels.append(higher)

                    elif nj < 0:
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][2][1]
                        else:
                            grids[higher] = zerogrid()
                            nextgrids[higher] = zerogrid()
                            newlevels.append(higher)

                    elif ni == len(grids[glevel]):
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][3][2]
                        else:
                            grids[higher] = zerogrid()
                            nextgrids[higher] = zerogrid()
                            newlevels.append(higher)

                    elif nj == len(grids[glevel][0]):
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][2][3]
                        else:
                            grids[higher] = zerogrid()
                            nextgrids[higher] = zerogrid()
                            newlevels.append(higher)
            
                if grids[glevel][i][j] == 1 and (not nsum == 1):
#                     print(glevel, i, j, nsum, 'dies')
                    nextgrids[glevel][i][j] = 0
                elif grids[glevel][i][j] == 0 and (nsum == 1 or nsum == 2):
#                     print(glevel, i, j, nsum, 'rises')
                    nextgrids[glevel][i][j] = 1
#                 if i == 0 and j == 1:
#                     print('catch', glevel, i, j, nsum, grids[glevel][i][j], nextgrids[glevel][i][j])
    
    for glevel in newlevels:
        for i in range(len(grids[glevel])):
            for j in range(len(grids[glevel][0])):
                nsum = 0
                
                if i == 2 and j == 2:
                    continue
                
                for adj in adjacents:
                    ni = i + adj[0]
                    nj = j + adj[1]
                
                    if ni == 2 and nj == 2:
                        lower = glevel + 1
                        if lower in grids:
                            if i == 1:
                                for lj in range(len(grids[lower][0])):
                                    nsum += grids[lower][0][lj]
                            elif i == 3:
                                for lj in range(len(grids[lower][0])):
                                    nsum += grids[lower][-1][lj]
                            elif j == 1:
                                for li in range(len(grids[lower])):
                                    nsum += grids[lower][li][0]
                            elif j == 3:
                                for li in range(len(grids[lower])):
                                    nsum += grids[lower][li][-1]

                    elif ni < 0:
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][1][2]

                    elif nj < 0:
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][2][1]

                    elif ni == len(grids[glevel]):
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][3][2]

                    elif nj == len(grids[glevel][0]):
                        higher = glevel - 1
                        if higher in grids:
                            nsum += grids[higher][2][3]

#                 print(glevel, i, j, nsum)
                if grids[glevel][i][j] == 1 and (not nsum == 1):
                    nextgrids[glevel][i][j] = 0
                elif grids[glevel][i][j] == 0 and (nsum == 1 or nsum == 2):
                    nextgrids[glevel][i][j] = 1
    return nextgrids
                    

def rate_biodiversity(grid):
    biodiverse = 0
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] == 1:
                exponent = i*len(grid[0]) + j
                biodiverse += 2**exponent
    return biodiverse
```

```python
rawgrid = """\
....#
#..#.
#..##
..#..
#....""".split('\n')

grid = parsegrid(rawgrid)
grids = {0: grid}

for i in range(200):
#     print('iter', i)
    grids = step_recur(grids)
    
bugsum = 0
for level in grids:
    bugsum += sum([sum(row) for row in grids[level]])
print(bugsum)

levels = list(grids.keys())
levels.sort()
# for level in range(-5, 6):
for level in levels:
    print(level)
    drawgrid(grids[level])
    print()
```

```python
rawgrid = """\
....#
#..#.
#..##
..#..
#....""".split('\n')

grid = parsegrid(rawgrid)
layouts = [grid]

i = 0
while True:
    grid = step(grid)
    if grid in layouts:
        print('found cycle in step', i)
        drawgrid(grid)
        bio = rate_biodiversity(grid)
        print('biodiversity', bio)
        break
    else:
        i += 1
        layouts.append(grid)
        
    if (i % 100) == 0:
        print('at cycle', i)
```
