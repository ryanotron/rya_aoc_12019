```python
def get_bestsite(astromap):
    astrolist = []
    viewmap = astromap.copy()
    for y in range(len(astromap)):
        for x in range(len(astromap[0])):
            if astromap[y][x] == '#':
                astrolist.append((x, y))
    
    viewcount = [0]*len(astrolist)
    for i, astro in enumerate(astrolist):
        xi = astro[0]
        yi = astro[1]
        
        for viewstro in astrolist:
            if viewstro == astro:
                continue
                
            xf = viewstro[0]
            yf = viewstro[1]
            
            dx = xf - xi
            dy = yf - yi
            
            obstructed = False
            for xinc in range(abs(dx)+1):
                for yinc in range(abs(dy)+1):
                    x = min([xi, xf]) + xinc
                    y = min([yi, yf]) + yinc
                    
                    if (x, y) == (xf, yf) or (x, y) == (xi, yi):
                        continue
                    
                    if (x, y) in astrolist:
                        err = x*dy - y*dx + yf*dx - xf*dy
                        if abs(err) < 1e-3:
                            obstructed = True
                            break
            if not obstructed:
                viewcount[i] += 1
    
    maxview = viewcount[0]
    idxmax = 0
    for i in range(0, len(viewcount)):
        vx = astrolist[i][0]
        vy = astrolist[i][1]
        
        viewmap[vy][vx] = str(viewcount[i])
        
        if viewcount[i] > maxview:
            maxview = viewcount[i]
            idxmax = i
            
    print('best site is ({}, {}) with {} views'\
          .format(astrolist[idxmax][0], astrolist[idxmax][1], viewcount[idxmax]))
    
    for row in viewmap:
        print(''.join(['{:4s}'.format(c) for c in row]))
```

```python
with open('calca_10_input.txt', 'r') as infile:
    rawstromap = [row.strip() for row in infile]
astromap = [[c for c in row] for row in rawstromap]

print(len(astromap), len(astromap[0]))
get_bestsite(astromap)
```

```python
def ccw2cw(rad):
    deg = int((rad*180.0/np.pi)//1)
    deg = (deg + 360) % 360
    deg -= 90
    if deg < 0:
        deg += 360
    return deg

def zaporder(astromap, centre=(0, 0)):
    zapx = centre[0]
    zapy = centre[1]
    
    zapmap = astromap.copy()
    zapmap[zapy][zapx] = 'X'
    
    astrolist = []
    for y in range(len(astromap)):
        for x in range(len(astromap[0])):
            if astromap[y][x] == '#':
                astrolist.append((x, y))
                
    print('there are {} asteroids'.format(len(astrolist)))
    zapspecs = []
    for aster in astrolist:
        asx = aster[0]
        asy = aster[1]
        
        dx = asx - zapx
        dy = asy - zapy
        
        if (abs(dx) < 1e-3) and (abs(dy) < 1e-3):
            print('skip', asx, asy)
            continue
        
        dis = np.sqrt(dx*dx + dy*dy)
        ang = np.arctan2(dy, dx)
        ang = 180*ang/np.pi
        ang = (ang + 90) % 360
        
        zapspecs.append([ang, dis, asx, asy])
        
    zapspecs.sort()
    
#     for i, zapspec in enumerate(zapspecs):
#         print('{:3d}: {:3d}, {:5.2f}, {:2d}, {:2d}'\
#               .format(i, zapspec[0], zapspec[1], zapspec[2], zapspec[3]))

    order = 1
    i = 0
    print('initial spec length', len(zapspecs))
    while len(zapspecs) > 0:
        x = zapspecs[i][2]
        y = zapspecs[i][3]
        r = zapspecs[i][0]
        w = zapspecs[i][1]
        print('zap {:3d}: ({:2d}, {:2d}, {:5.2f}, {:5.2f})'.format(order, x, y, r, w), len(zapspecs))
        zapmap[y][x] = str(order)
        order += 1
        theta = zapspecs[i][0]        
        zapspecs = zapspecs[:i] + zapspecs[i+1:]
        if len(zapspecs) < 1:
            break
        
        j = i
        if j >= len(zapspecs):
            j = 0
        
        while zapspecs[j][0] == theta:
            j += 1
            if j >= len(zapspecs):
                j = 0
                theta = -360   
        i = j
        
    for row in zapmap:
        print(''.join(['{:4s}'.format(c) for c in row]))
```
